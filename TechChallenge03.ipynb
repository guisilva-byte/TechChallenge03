```python
data = []

with open(uncompressed_path, 'r', encoding='utf-8') as file:
    for line_number, line in enumerate(file, 1):
        line = line.strip()
        if line:
            try:
                record = json.loads(line)
                data.append(record)
            except json.JSONDecodeError as e:
                print(f"Erro ao decodificar JSON na linha {line_number}: {e}")
                break

print(f"Total de registros carregados: {len(data)}")
```

```python
from unsloth import FastLanguageModel, is_bfloat16_supported

model_name = "unsloth/llama-3-8b-bnb-4bit"
load_in_4bit = True
max_length = 2048

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name,
    load_in_4bit=load_in_4bit,
    device_map="auto"
)
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16,
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0,
    bias = "none",

    use_gradient_checkpointing = "unsloth",
    random_state = 3407,
    use_rslora = False,
    loftq_config = None,
)
```

```python
FastLanguageModel.for_inference(model)
```

```python
test_samples = test_dataset.select(range(20))
```

```python
def generate_response(model, tokenizer, input_question, max_length=2048):
    instruction = "Please answer the question to the best of your ability."
    prompt = f"{instruction}\n\nQuestion:\n{input_question}\n\nAnswer:"

    # Tokenizar a entrada
    inputs = tokenizer(
        prompt,
        return_tensors='pt',
        truncation=True,
        max_length=max_length
    )

    # Mover para o dispositivo apropriado
    first_device = next(model.parameters()).device
    inputs = {key: value.to(first_device) for key, value in inputs.items()}

    # Gerar a sada
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=150,
            do_sample=True,
            temperature=0.7,
            top_p=0.9,
            num_return_sequences=1
        )

    # Decodificar a sada
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    generated_response = generated_text[len(prompt):].strip()

    return generated_response
```

```python
for sample in test_samples:
    input_question = sample['text']  # Ou use uma pergunta especfica
    response = generate_response(model, tokenizer, input_question)
    print(f"Question: {input_question}\nAnswer: {response}\n")
```

```python
# Funo de pr-processamento
def preprocess_function(examples):
    instruction = "Please answer the question to the best of your ability."
    inputs = [
        f"{instruction}\n\nQuestion:\n{txt}\n\nAnswer:"
        for txt in examples["text"]
    ]
    model_inputs = tokenizer(
        inputs,
        max_length=max_length,
        truncation=True,
    )

    # Os labels sero os mesmos que os inputs (modelo aprende a gerar a sequncia completa)
    model_inputs["labels"] = model_inputs["input_ids"].copy()
    return model_inputs

# Aplicar a funo ao conjunto de treinamento
tokenized_train_dataset = train_dataset.map(preprocess_function, batched=True, remove_columns=train_dataset.column_names)
```

```python
from trl import SFTTrainer
from transformers import TrainingArguments

max_seq_length = 2048
is_bf16_supported = is_bfloat16_supported()

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=tokenized_train_dataset,
    dataset_text_field="input_ids",
    max_seq_length=max_seq_length,
    dataset_num_proc=2,
    packing=False,
    args=TrainingArguments(
        per_device_train_batch_size=1,  # Ajuste conforme necessrio
        gradient_accumulation_steps=4,
        warmup_steps=5,
        max_steps=60,
        learning_rate=2e-4,
        fp16=not is_bf16_supported,
        bf16=is_bf16_supported,
        logging_steps=1,
        optim="adamw_8bit",
        weight_decay=0.01,
        lr_scheduler_type="linear",
        seed=3407,
        output_dir="outputs",
    ),
)
```

```python
trainer.train()
```

```python
# Salvar o modelo treinado
trainer.save_model("fine_tuned_model")
tokenizer.save_pretrained("fine_tuned_model")
```

```python
model_ft, tokenizer_ft = FastLanguageModel.from_pretrained("fine_tuned_model")

# Preparar o modelo para inferncia
FastLanguageModel.for_inference(model_ft)
```

```python
def generate_response_ft(model, tokenizer, input_text, max_length=2048):
    instruction = "Por favor, responda  pergunta da melhor forma possvel."
    prompt = f"{instruction}\n\nPergunta:\n{input_text}\n\nResposta:"

    # Tokenizar a entrada
    inputs = tokenizer(
        prompt,
        return_tensors='pt',
        truncation=True,
        max_length=max_length
    )

    # Obter o dispositivo da primeira camada do modelo
    first_device = next(model.parameters()).device
    inputs = {key: value.to(first_device) for key, value in inputs.items()}

    # Gerar a sada
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=150,
            do_sample=True,
            temperature=0.7,
            top_p=0.9,
            num_return_sequences=1
        )

    # Decodificar a sada
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    generated_response = generated_text[len(prompt):].strip()

    return generated_response
```
